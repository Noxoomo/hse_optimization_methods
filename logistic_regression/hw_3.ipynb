```python
from logistic_regression.oracle import Oracle
from logistic_regression.optimizers import GradientDescent
import numpy as np
import plotly.graph_objs as go
```

```python
oracle = Oracle.make_oracle("data/a1a.txt", "libsvm")
optimizer = GradientDescent(oracle, "nesterov")
```

```python
# np.random.seed(42)
w_0 = np.zeros(oracle._x.shape[1], dtype=np.float64)
w_opt = optimizer(w_0, max_iter=10000)
```

```python
stats = optimizer.stats
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=stats["calls"], y=stats["errors"], mode='lines'))
fig.update_layout(
    yaxis={"title": "r_k errors", "exponentformat": "e", "type": "log"},
    xaxis={"title": "Oracle calls"},
    title=f"Error dependency by Oracle calls"
)
fig.show()
```
